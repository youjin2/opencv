## basic operators
* cv clips pixels exceeding 0-255 to 0-255
    * c.f. numpy does not
    * e.g.
        - cv: 200 + 100 = 255
        - np: 200 + 100 = 44
* see
    * cv.add
    * cv.subtract
    * cv.multiply
    * cv.divide

```python
img1 = cv.imread('../img/wing_wall.jpg')
img2 = cv.imread('../img/yate.jpg')
```

```python
img3 = img1 + img2
img4 = cv.add(img1, img2)
```

```python
display_image(img1)
```

```python
display_image(img2)
```

```python
display_image(img3)
```

```python
display_image(img4)
```

## alpha blending

```python
alpha = 0.5
```

```python
dst = cv.addWeighted(img1, alpha, img2, (1-alpha), 0)
```

```python
display_image(dst)
```

## bitwise-operators
* 255=1, 0=0
* see
    * cv.bitwise_and
    * cv.bitwise_or
    * cv.bitwise_xor
    * cv.bitwise_not

```python
img = cv.imread('../img/girl.jpg')
```

```python
mask = np.zeros_like(img)
```

```python
masked = cv.bitwise_and(img, mask)
```

```python
display_image(img)
```

```python
display_image(mask)
```

```python
display_image(masked)
```
