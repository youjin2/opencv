## gray scale

```python
img = cv.imread('../img/girl.jpg')
```

```python
img2 = img.astype(np.uint16)
```

```python
b, g, r = cv.split(img2)
```

* may exceeds 255 in calculating b+g+r

```python
gray1 = ((b + g + r) / 3).astype(np.uint8)
```

```python
gray2 = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

* cv img is BGR order
    - one should convert gbr to rgb to display image using matplotlib (RGB)

```python
plt.imshow(img)
```

```python
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
```

```python
plt.imshow(gray1, cmap='gray')
```

```python
plt.imshow(gray2, cmap='gray')
```
